```python
import os
import cv2
root = r'C:\Users\Rong_\Documents\data\data_DTU'
for scan in os.listdir(root):
    scan = os.path.join(root, scan)
    image_folder = os.path.join(scan, 'image')
    mask_folder = os.path.join(scan, 'mask')
    mask_image_folder = os.path.join(scan, 'mask_image')
    os.makedirs(mask_image_folder, exist_ok=True)
    for image_idx, image_name in enumerate(os.listdir(image_folder)):
        image_path = os.path.join(image_folder, image_name)
        mask_path = os.path.join(mask_folder, f'{image_idx}'.zfill(3) + '.png')

        # Read the image and the mask
        image = cv2.imread(image_path)
        mask = cv2.imread(mask_path, 0)  # 0 to read in grayscale

        if image is not None and mask is not None:
            # Apply the mask
            masked_image = cv2.bitwise_and(image, image, mask=mask)

            # Save the masked image
            save_path = os.path.join(mask_image_folder, f'{image_idx}'.zfill(6) + '.png')
            cv2.imwrite(save_path, masked_image)
        else:
            print(f"Image or mask not found for index {image_idx}")
```

```python
def rotmat2qvec(R):
    Rxx, Ryx, Rzx, Rxy, Ryy, Rzy, Rxz, Ryz, Rzz = R.flat
    K = np.array([
        [Rxx - Ryy - Rzz, 0, 0, 0],
        [Ryx + Rxy, Ryy - Rxx - Rzz, 0, 0],
        [Rzx + Rxz, Rzy + Ryz, Rzz - Rxx - Ryy, 0],
        [Ryz - Rzy, Rzx - Rxz, Rxy - Ryx, Rxx + Ryy + Rzz]]) / 3.0
    eigvals, eigvecs = np.linalg.eigh(K)
    qvec = eigvecs[[3, 0, 1, 2], np.argmax(eigvals)]
    if qvec[0] < 0:
        qvec *= -1
    return qvec
import os
import numpy as np
root = r'C:\Users\Rong_\Documents\data\data_dtu'
for scan in os.listdir(root):
    scan = os.path.join(root, scan)
    print(scan)
    camera = np.load(os.path.join(scan, 'cameras_large.npz'))
    sparse_path = os.path.join(scan, 'sparse'+os.sep+'0')
    image = ''
    for i in range(len(camera.keys())//6):
        print(str(i).zfill(6)+'.png')
        break

    


    Rt = camera['world_mat_0']
    S = camera['scale_mat_0']
    K = camera['camera_mat_0']
    R = Rt[:3,:3]
    t = Rt[:3,3]
    q = rotmat2qvec(R)
    print(q)
    os.makedirs(sparse_path, exist_ok=True)
    with open(os.path.join(os.path.join(scan, 'sparse'+os.sep+'0'), 'cameras.txt'),'w')as f:
        f.write(K.tostring())
    with open(os.path.join(os.path.join(scan, 'sparse'+os.sep+'0'), 'images.txt'),'w')as f:
        f.write('')
    with open(os.path.join(os.path.join(scan, 'sparse'+os.sep+'0'), 'points3D.txt'),'w')as f:
        f.write('')
    break
```

```python
import os
import shutil

root = r'C:\Users\Rong_\Documents\data\data_DTU'
des = r'C:\Users\Rong_\Documents\data\DTU'

for scan in os.listdir(root):
    scan_path = os.path.join(root, scan)
    des_scan_path = os.path.join(des, scan)

    # Create corresponding scan directory in the destination
    if not os.path.exists(des_scan_path):
        os.makedirs(des_scan_path)

    # Check and copy the mask_image folder
    mask_image_path = os.path.join(scan_path, 'mask_image')
    if os.path.exists(mask_image_path):
        des_image_path = os.path.join(des_scan_path, 'images')
        shutil.copytree(mask_image_path, des_image_path)

    # Check and copy the cameras_large.npz file
    cameras_file_path = os.path.join(scan_path, 'cameras_large.npz')
    if os.path.isfile(cameras_file_path):
        des_cameras_file_path = os.path.join(des_scan_path, 'cameras.npz')
        shutil.copy2(cameras_file_path, des_cameras_file_path)
```
